```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import requests
from bs4 import BeautifulSoup
import re
```

2. Instantiate Model

```python
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')

model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

3. Encode and Calculate Sentiment

```python
tokens = tokenizer.encode('It was good but couldve been better. Great', return_tensors='pt')
```

```python
result = model(tokens)
```

```python
result.logits
```

```python
int(torch.argmax(result.logits))+1
```

Scraping from MAL

```python
r = requests.get('https://myanimelist.net/anime/1735/Naruto__Shippuuden/reviews')
```

```python
r.text
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
regex = re.compile('.*text.*')
review_results = soup.find_all('div', {'class': regex})
reviews = [result.text.strip() for result in review_results]
```

```python
# Find usernames
username_results = soup.find_all('div', class_='username')
usernames = [user.a.text for user in username_results]
```

```python
usernames
```

```python
reviews
```

5. Load Reviews into Dataframe and Score

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'username': usernames, 'review': reviews})
```

```python
df
```

```python
def sentiment_score(review):
    tokens = tokenizer.encode(review, return_tensors='pt', truncation=True, max_length=512)
    result = model(tokens)
    return int(torch.argmax(result.logits)) + 1  # Sentiment score (1 to 5)
```

```python
df['sentiment'] = df['review'].apply(lambda x: sentiment_score(x))
```

```python
# Tambahkan gaya untuk mempercantik DataFrame
def style_table(val):
    if val == 5:
        return 'background-color: lightgreen; color: black;'
    elif val == 1:
        return 'background-color: lightcoral; color: white;'
    else:
        return ''

# Terapkan gaya pada DataFrame
styled_df = df.style.map(style_table, subset=['sentiment']) \
                    .set_properties(**{'text-align': 'left'}) \
                    .set_table_styles([{
                        'selector': 'th',
                        'props': [('text-align', 'left')]
                    }])

# Tampilkan DataFrame tanpa indeks (versi manual)
styled_df = styled_df.format_index(lambda x: "", axis=0)  # Menyembunyikan indeks
styled_df
```
